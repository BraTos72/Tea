```python
# Analizar el crecimiento de la poblacion Ecuador y Honduras
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
datos = pd.read_csv("countries.csv")
```

```python
datos
```

```python
ecuador= datos[datos.country =="Ecuador"]
```

```python
ecuador
```

```python
honduras = datos[datos.country == "Honduras"]
```

```python
honduras
```

```python
plt.plot(ecuador.year, ecuador.population/10**6)
plt.plot(honduras.year, honduras.population/10**6)
plt.legend(["Ecuador","Honduras"])
plt.xlabel("Ao")
plt.ylabel("Poblacipon(millones)")
plt.show()
```

```python
ecuador.population
```

```python
ecuador.population / ecuador.population.iloc[0]
```

```python
ecuador.population / ecuador.population.iloc[0]*100
```

```python
plt.plot(ecuador.year, ecuador.population / ecuador.population.iloc[0]*100)
plt.plot(honduras.year, honduras.population / honduras.population.iloc[0]*100)
plt.legend(["Ecuador","Honduras"])
plt.xlabel("Ao")
plt.ylabel("Crecimiento de la poblacin (1er ao 100%)")
plt.show()
```
